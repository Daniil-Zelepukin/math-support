```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1*np.pi, 5)
y = np.sin(x)

plt.plot(x, y)
plt.title('График построения косинуса')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()
```
